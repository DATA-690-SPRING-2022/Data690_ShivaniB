```python
df_ind = df_new[df_new['Country Code']== 'IND']
df_ind
```

## Calculating the number of infant deaths in india

```python
df_ind_infant = df_ind.query("indicator in(['SH.DTH.IMRT'])")
df_ind_infant
```

```python
df_pivot = df_ind_infant.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "SH.DTH.IMRT",
    title = "Infant deaths",
    labels = {"SH.DTH.IMRT":"infant male deaths ","SH.DTH.IMRT":"total child deaths"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    #size = "SP.POP.TOTL"
)
#fig.update_traces(textposition = "top center")

fig.show()
```

## Educational attainment in the selected countries

```python
df_edu =  df_new[df_new["Country Code"].isin(["DEU","KOR","PAK","USA"])]
df_edu
```

```python
df_edu_country = df_edu.query("indicator in(['SE.PRM.CUAT.ZS'])")
df_edu_country
```

```python
df_pivot = df_edu_country.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "SE.PRM.CUAT.ZS",
    title = "Education enrollment",
    labels = {"Year":"year","SE.PRM.CUAT.ZS":"educational attainment"},
    color = "Country Name",
    animation_frame = "Year",
    log_x=False,
    animation_group = "Country Name"
    #size = "SP.POP.TOTL"
)

#fig.update_traces(textposition = "top center")

fig.show()
```

## Retirement benefits

```python
df_retirement = df_new.query("indicator in(['SG.AGE.RTRE.PL.FE','SG.AGE.RTRE.PL.MA'])")
df_retirement
```

```python
df_pivot_ret = df_retirement.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot_ret
```

```python
fig = px.pie(data_frame = df_pivot_ret, values='SG.AGE.RTRE.PL.FE', names='Country Name',title = "retirement benefits for women", hover_name= "Country Name" ,color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()

fig = px.pie(data_frame = df_pivot_ret, values='SG.AGE.RTRE.PL.MA', names='Country Name',title = "retirement benefits for men", hover_name= "Country Name" ,color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
